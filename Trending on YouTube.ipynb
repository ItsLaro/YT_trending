# Trending on YouTube

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
import seaborn as sns
%matplotlib inline
```

```python
youtube = pd.read_csv('USvideos.csv') #Method reads the info for the CSV (Comma Separated Values)
youtube
```

```python
youtube.info()
```

```python
youtube.describe()
```

```python
channel_views = youtube.groupby('channel_title').sum()
channel_views.sort_values('views', ascending = False).head(n=10)
```

```python
youtube['net like-dislike'] = youtube['likes'] - youtube['dislikes']
youtube.head()
```

```python
youtube = youtube[(youtube['comments_disabled'] == False) & (youtube['ratings_disabled'] == False)]
youtube
```

```python
sns.lmplot(x='net like-dislike', y='comment_count', data=youtube, aspect=2, height=10)
```
